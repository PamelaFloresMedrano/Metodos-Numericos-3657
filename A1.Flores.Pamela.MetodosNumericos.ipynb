#### 1. Desarrolle en Python un programa para calcular la inversa de matrices de dimensión 2×2, no olvide colocar comentarios en su programa. No busque el programa en internet.

```python
# Importamos el paquete de numpy
import numpy as np

# Creamos una matriz 2x2 de enteros aleatorios
A = np.random.randint(10, size=(2,2))
print("La matriz de 2x2 generada es:\n")
print(A)

#Calculamos el determinante para validar que se pueda calcular la inversa
#Inicializamos en 0
det=0.0
det = np.linalg.det(A)
print("\nEl determinante de la matriz 2x2 es:")
print(int(det))

#Verificamos que el determinante no sea cero
if (det)!= 0 :
    inversa_A = np.linalg.inv(A)
    print("\nLa matriz inversa es:\n")
    print(inversa_A)
else:
    print("\nNo se puede encontrar la matriz inversa")
```

#### 2. Grafique en Python las siguientes funciones, $f(x)= x^2−x+1$, $g(x)=\frac{2}{x-1}$. Grafique ambas funciones en el mismo gráfico.

```python
#Importamos la librería matplotlib
import matplotlib.pyplot as plt
import numpy as np

#Se muestra la salida de los comandos en línea
%matplotlib inline

# Función para evaluar f(x) y g(x) en un rango largo
x = np.linspace(-5,5,100)

# Creamos las funciones solicitadas
def f(x):
    return ((x**2)-x+1)
def g(x):
    return (2/(x-1))

#Función que crea la figura
fig, ax = plt.subplots()

#Funciones que crean el entorno gráfico
ax.plot(x,f(x))
ax.plot(x,g(x))
ax.set_title('Gráfico de Funciones f(x) y g(x)')

# Codigo para mostrar ejes de coordenadas dentro del gráfico creado
xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()

ax.grid(True, linestyle='-')
ax.annotate("",xy=(xmax, 0), xytext=(xmin, 0), 
            arrowprops=dict(color='black',width=1.5, headwidth=8, headlength=10))
ax.annotate("",xy=(0, ymax), xytext=(0, ymin), 
            arrowprops=dict(color='black',width=1.5, headwidth=8, headlength=10))

# Comando que muestra la figura resultante
plt.show()
```

#### 5. Diseñe un código que encuentre el $sen (\frac{π}{3})$ a través del desarrollo de Taylor, truncar cuando $n = 50$.

Tenemos en cuenta que la función seno aproximada con la serie de Taylor viene definida por:

$sen(x)= \sum _{k=0}^{\infty }\:\left(-1\right)^k\frac{\:x^{\left(2k+1\right)}}{\left(2k+1\right)!}$

```python
#Importamos la librería math
import math

#Transformamos de grados a radianes haciendo uso de la librería
xgrad = float(input("Cual es el valor en grados del seno: "))
x = math.radians(xgrad)
print("El valor en radianes es:",x)
```

```python
# Digitamos los términos a sumar
n = int(input("Numero de términos a sumar: "))

#Inicializa en cero
sen_x=0.0

#Imprimimos en pantalla
print("\n\n{:^2} {:^10} {:^16}".format("Término", "k","sen(π/3)\n"))

# Hacemos uso del bucle for para iterar/sumar los términos solicitados
for k in range(n):
    sen_x = sen_x + (-1)**k * x**(2*k+1)/math.factorial((2*k+1))
    print("{:^2d} {:^15d} {:^12.10f}".format(k+1, k,sen_x))
```
