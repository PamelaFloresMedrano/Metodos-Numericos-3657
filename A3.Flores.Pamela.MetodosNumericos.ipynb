**3. Utilice el método de Jacobi, Gauss-Seidel y SOR (ω = 1.1) para resolver el siguiente sistema lineal con una precisión de cuatro cifras decimales.**


<center>$\begin{pmatrix}7&1&-1&2\\ 1&8&0&-2\\ -1&0&4&-1\\ 2&-2&-1&6\end{pmatrix}\begin{pmatrix}x_1\\ x_2\\ x_3\\ x_4\end{pmatrix}=\begin{pmatrix}3\\ -5\\ 4\\ -3\end{pmatrix}$
    
**Compare el número de iteraciones necesario en cada algoritmo. Analice el error cometido si la solución exacta es $x\:=\:\left(1,\:−1,\:1,\:−1\right)^T$**


**MÉTODO DE JACOBI**

```python
import numpy as np

A = np.array([[ 7, 1,-1,  2], 
              [ 1, 8, 0, -2], 
              [-1, 0, 4, -1],
              [ 2,-2, -1, 6]])

b = np.array([3, -5, 4, -3])

x0 = np.zeros(len(b))

#Comprobamos que la matriz de entrada es diagonal dominante
def diag_dom(X):
    
    #Encontramos los coeficientes de la diagonal
    D = np.diag(np.abs(X))
    S = np.sum(np.abs(X), axis=1) - D 
    
    if np.all(D > S):
        print("La matriz SI es diagonal dominante\n")
    else:
        print("La matriz NO es diagonal dominante\n")
    return

diag_dom(A)

#Implementamos la función para el método de Jacobi
def Jacobi(A, b, x0, kmax = 100, tol = 1e-4):
    #Utilizamos el método copy para que la lista original no cambie cuando se modifique la nueva lista en el método.
    x_new = x0.copy() 
    #Usamos la función numpy.shape() el cual encuentra la longitud de las dimensiones del array A.
    fil, col = np.shape(A)
    
    #Método de Jacobi
    for k in range(kmax): 
        x_copy = x_new.copy()
        for i in range(fil):     
            sum = 0        
            for j in range(col):
                if j != i:
                    #Fórmula de iteración
                    sum += A[i, j] * x_new[j]
            x_copy[i] = (b[i] - sum) / A[i][i]
        #Diferencia con Gauss-Seidel
        x_new = x_copy.copy()
        
        # Criterio de parada
        Ax = np.dot(A, x_new)               
        cp = np.linalg.norm(b - Ax)
        print("Iteración", k+1,"°",'Residuo: {0:10.6g}'.format(cp))
        if cp < tol:
            print ("\nEl número máximo de iteraciones alcanzado es: ", k+1, " iteraciones")
            break
    return x_new

#Imprimimos resultados
x = Jacobi(A, b, x0)   
print("\nLa solución por el Método de Jacobi es:\n", x)
```

**MÉTODO GAUSS-SEIDEL**

```python
import numpy as np

A = np.array([[ 7, 1,-1,  2], 
              [ 1, 8, 0, -2], 
              [-1, 0, 4, -1],
              [ 2,-2, -1, 6]])
                  
b = np.array([3, -5, 4, -3])

x0 = np.zeros(len(b)) 

#Comprobamos que la matriz de entrada es diagonal dominante
def diag_dom(X):
    
    #Encontramos los coeficientes de la diagonal
    D = np.diag(np.abs(X))
    S = np.sum(np.abs(X), axis=1) - D 
    
    if np.all(D > S):
        print("La matriz SI es diagonal dominante\n")
    else:
        print("La matriz NO es diagonal dominante\n")
    return 

diag_dom(A)

def GaussSeidel(A, b, x0, kmax = 100, tol = 1e-4):
    #Utilizamos el método copy para que la lista original no cambie cuando se modifique la nueva lista en el método.
    x_new = x0.copy()
    #Usamos la función numpy.shape() el cual encuentra la longitud de las dimensiones del array A.
    fil, col = np.shape(A)
    
    #Método de Gauss-Seidel
    for k in range(kmax): 
        for i in range(fil):     
            sum = 0        
            for j in range(col):
                if j != i:
                    #Fórmula de iteración
                    sum += A[i, j] * x_new[j]
            x_new[i] = (b[i] - sum) / A[i][i]
            
 # Criterio de parada
        Ax = np.dot(A, x_new)               
        cp = np.linalg.norm(b - Ax)
        print("Iteración",k+1,"°",'Residuo: {0:10.6g}'.format(cp))
        if cp < tol:
            print ("\nEl número máximo de iteraciones alcanzado es: ", k+1, " iteraciones")
            break
        
    return x_new

#Imprimimos resultados
x = GaussSeidel(A, b, x0)
print("\nLa solución por el Método de Gauss Seidel es:\n", x)
```

**MÉTODO SUCCESSIVE OVER RELAXATION (SOR)**

```python
import numpy as np

A = np.array([[ 7, 1,-1,  2], 
              [ 1, 8, 0, -2], 
              [-1, 0, 4, -1],
              [ 2,-2, -1, 6]])
                  
b = np.array([3, -5, 4, -3])

x0 = np.zeros(len(b)) 

#Comprobamos que la matriz de entrada es diagonal dominante
def diag_dom(X):
    
    #Encontramos los coeficientes de la diagonal
    D = np.diag(np.abs(X))
    S = np.sum(np.abs(X), axis=1) - D 
    
    if np.all(D > S):
        print("La matriz SI es diagonal dominante\n")
    else:
        print("La matriz NO es diagonal dominante\n")
    return

diag_dom(A)

def SOR(A, b, x0, omega=1.1, kmax = 100, tol = 1e-4):
    #Utilizamos el método copy para que la lista original no cambie cuando se modifique la nueva lista en el método.
    x_new = x0.copy()
    #Usamos la función numpy.shape() el cual encuentra la longitud de las dimensiones del array A.
    fil, col = np.shape(A)
    
    #Método SOR
    for k in range(kmax): 
        for i in range(fil):     
            sum = 0        
            for j in range(col):
                if j != i:
                    #Fórmula de iteración
                    sum += A[i, j] * x_new[j]
            x_new[i] = (1 - omega) * x_new[i] + (omega / A[i][i]) * (b[i] - sum)
            
# Criterio de parada
        Ax = np.dot(A, x_new)               
        cp = np.linalg.norm(b - Ax)
        print("Iteración",k+1,"°",'Residuo: {0:10.6g}'.format(cp))
        if cp < tol:
            print ("\nEl número máximo de iteraciones alcanzado es: ", k+1, " iteraciones")
            break
        
    return x_new

x = SOR(A, b, x0)
    
print("\nLa solución por el Método de SOR es:\n", x)
```

**4. Programe un algoritmo para encontrar la norma de Frobenius para una matriz cuadrada de cualquier dimensión.**


```python
import numpy as np

#Pedimos por pantalla la dimensión de la matriz
print("Introduzca la dimensión de la matriz cuadrada:")
n = int(input())

#Creamos una matriz cuadrada randómica
A = np.random.random((n,n))

#Algoritmo Norma de Frobenius
def norma_frobenius(A):
    sum = 0
    for i in A:
        for elem in i:
            sum = sum + elem**2
    sum = sum**(1 / 2)
    print("La matriz es:\n", A)
    return sum

print("\nLa norma de Frobenius para la matriz cuadrada dada es: ",norma_frobenius(A))
```
