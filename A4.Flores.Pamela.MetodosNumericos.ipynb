**3. Una compañía fabrica dos productos, A y B. Los ingresos unitarios son 2 y 3 dólares, respectivamente. Las disponibilidades diarias de dos materias primas, M1 y M2, utilizadas en la fabricación de los dos productos son de 8 y 18 unidades, respectivamente. Una unidad de A utiliza 2 unidades de M1 y 2 unidades de M2, y una unidad de B utiliza 3 unidades de M1 y 6 unidades de M2.**

Tabla de Datos:



Tomamos a $x_1$ como la cantidad de producto A y $x_2$ la cantidad de producto B que se van a fabricar y además dos restricciones de disponibilidad de materia prima M1 y M2. 

**Función Objetivo:** $z = 2x_1 + 3x_2$

Planteamos el problema en notación de programación lineal:

**<center>Maximizar $z= 2x_1 + 3x_2$</center>** 

**<center>sujeto a $2x_1 + 2x_2 ≤ 8$</center>**

**<center>$3x_1 + 6x_2 ≤ 18$</center>**

**<center>$x_1, x_2 ≥ 0$</center>**

**(a) Determine los precios duales de M1 y M2 y sus intervalos de factibilidad.**

Si resolvemos de forma gráfica obtenemos el óptimo en $x_1 = 2$, $x_2 = 2$ con un valor de $z=10$


Si a la disponibilidad de la materia prima le aumentamos una unidad, en este caso a M1, tenemos el siguiente problema:


**<center>Maximizar $z= 2x_1 + 3x_2$</center>** 

**<center>sujeto a $2x_1 + 8x_2 ≤ 9$</center>**

**<center>$3x_1 + 6x_2 ≤ 18$</center>**

**<center>$x_1, x_2 ≥ 0$</center>**

Si volvemos a resolver este nuevo problema de forma gráfica obtenemos el óptimo en $x_1 = 3$, $x_2 = 1.5$ con un valor de $z=10.5$



Definimos el precio dual como: 
$\frac{valor\:óptimo\:nuevo\:-\:valor\:óptimo\:anterior}{capacidad\:actual\:-\:capacidad\:anterior}=\frac{z_B-z_A}{9-8}$

$=\frac{10.5-10}{9-8}=\$0.50\:/ unidad$

Este resultado nos quiere decir que si aumentamos la disponibilidad de la materia prima M1 en una unidad tendremos $\$0.50\:$ más de ingresos y si disminuimos la disponibilidad en uno de la materia prima M1 tendremos una disminución de ganancia de $\$0.50\:$.

**<center>Maximizar $z= 2x_1 + 3x_2$</center>**

**<center>$3x_1 + 6x_2 ≤ 19$</center>**

**<center>$x_1, x_2 ≥ 0$</center>**

Si volvemos a resolver este nuevo problema de forma gráfica obtenemos el óptimo en $x_1 = 1.67$, $x_2 = 2.33$ con un valor de $z=10.33$



Definimos el precio dual como: 
$\frac{valor\:óptimo\:nuevo\:-\:valor\:óptimo\:anterior}{capacidad\:actual\:-\:capacidad\:anterior}=\frac{z_B-z_A}{19-18}$

$=\frac{10.33-10}{19-18}=\$0.33\:/ unidad$

Si comparamos los precios duales podemos concluir que obtendremos más ganancia al aumentar la disponibilidad de la materia prima M1.
